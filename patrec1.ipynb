```python
def digit_mean_at_pixel(X, y, digit, pixel=(10, 10)):
   
    li = []
    for i in range(len(y)):
        if y[i]==digit :
           
            aaa=np.reshape(X[i],(16,16))
            
            li.append(aaa[pixel[0]][pixel[1]])
            
    
    return np.mean(li)
            
mean0=  digit_mean_at_pixel(X_train, y_train, 0, pixel=(10, 10))   
print(mean0)   
```

```python
def digit_variance_at_pixel(X, y, digit, pixel=(10, 10)):
   
    li = []
    for i in range(len(y)):
        if y[i]==digit :
            #c+=1
            
            aaa=np.reshape(X[i],(16,16))
            
            li.append(aaa[pixel[0]][pixel[1]])
            
    return np.var(li)

variance0=digit_variance_at_pixel(X_train, y_train, 0, pixel=(10, 10))
print(variance0)
    
```

```python
def digit_mean(X, y, digit):
    mean=np.zeros((16,16))
    for i in range(16):
        for j in range(16):
            mean[i][j]=digit_mean_at_pixel(X, y, digit, pixel=(i, j))
    return mean

mean0new=digit_mean(X_train,y_train,0)
print(mean0new)
```

```python
def digit_variance(X, y, digit):
    var=np.zeros((16,16))
    for i in range(16):
        for j in range(16):
            var[i][j]=digit_variance_at_pixel(X, y, digit, pixel=(i, j))
    return var

variance0 = digit_variance(X_train,y_train,0)
print(variance0)
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_X_y, check_is_fitted
from sklearn.utils.multiclass import unique_labels

class EuclideanDistanceClassifier(BaseEstimator, ClassifierMixin):
    """Classify samples based on the distance from the mean feature value"""

    def __init__(self,X_mean_=None):
        self.X_mean_ = None


    def fit(self, X, y):
        """
        This should fit classifier. All the "work" should be done here.
        Calculates self.X_mean_ based on the mean
        feature values in X for each class.
        self.X_mean_ becomes a numpy.ndarray of shape
        (n_classes, n_features)
        fit always returns self.
        """
        X, y = check_X_y(X, y)
        
        self.classes_ = unique_labels(y)
        self.X_ = X
        self.y_ = y
        
        n_classes = self.classes_.shape[0]
        #print(n_classes)
        n_features=X.shape[1]
        #print(n_features)
        self.X_mean_ = np.zeros((n_classes,n_features))
        for i in range(n_classes):
             self.X_mean_[i] = digit_mean(X,y,i).ravel() 
        
        
        
        
        
        
        
        return self
      


    def predict(self, X):
        """
        Make predictions for X based on the
        euclidean distance from self.X_mean_
        """
        check_is_fitted(self)
        predictions = euclidean_distance_classifier(X,self.X_mean_)
        
        #print(X.shape)
        #print(self.X_mean_.shape)
        return predictions
        #EuclideanDistanceClassifier(X,self.X_mean_)

    def score(self, X, y):
        """
        Return accuracy score on the predictions
        for X based on ground truth y
        """
        accurate=0
        estimated=self.predict(X)
        for i in range(len(estimated)):
            if estimated[i] == y[i]:
                accurate+=1
                suc_rate = accurate/len(estimated)*100


        return round(suc_rate,4)

from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
def evaluate_classifier(clf, X, y, folds=5):
       
#     print("score is" + str(clf.score(X,y)*100))
    scores = cross_val_score(clf, X, y, 
    cv=KFold(n_splits=5, random_state=None),scoring="accuracy")
    print("CV error = %f +-%f" % (1. - np.mean(scores), np.std(scores)))
    
```

```python
def calculate_priors(X, y):
    """Return the a-priori probabilities for every class
    Args:
        X (np.ndarray): Digits data (nsamples x nfeatures)
        y (np.ndarray): Labels for dataset (nsamples)
    Returns:
        (np.ndarray): (n_classes) Prior probabilities for every class
    """

    labels = np.unique(y.ravel())
    labels = labels.astype(int)
    #print(labels)
    
    priors = np.zeros(len(labels))
    
    for label in labels:
        #print(label)
        priors[label] = np.count_nonzero(y==label)/len(y)
        #print(priors[label])
    return priors

priors=calculate_priors(X_train,y_train)
print(priors)
```

```python
import math
class CustomNBClassifier(BaseEstimator, ClassifierMixin):
    """Custom implementation Naive Bayes classifier"""

    def __init__(self, use_unit_variance=False):
        self.use_unit_variance = use_unit_variance


    def fit(self, X, y):
        X, y = check_X_y(X, y)
        
        self.classes_ = unique_labels(y)
        self.X_ = X
        self.y_ = y
        self.priors = calculate_priors(X,y)
        
        n_classes = self.classes_.shape[0]
        
        n_features=X.shape[1]
        
        n_data= X.shape[0]
        
        self.X_mean_ = np.zeros((n_classes,n_features))
        self.X_var_ = np.ones((n_classes,n_features))
        self.X_prob_ = np.zeros(n_data)
        for i in range(n_classes):
            self.X_mean_[i]=digit_mean(X,y,i).ravel()
            if(not self.use_unit_variance):
                self.X_var_[i] = digit_variance(X,y,i).ravel()
        
        
#         print(self.X_mean_)
#         print(self.X_var_)
        
        
        return self


    def predict(self, X):
        """
        Make predictions for X based on the
        euclidean distance from self.X_mean_
        """
        predictions=np.zeros(X.shape[0])
        for i in range(X.shape[0]):
            m=-sys.maxsize
            for j in range(len(self.X_mean_)):
                prod = 1
                for k in range(X.shape[1]):
                    prod*=((1/math.sqrt(2*np.pi*self.X_var_[j][k]**2+0.0000000001)*math.exp(-(X[i][k]-self.X_mean_[j][k])**2/(2*self.X_var_[j][k]**2+0.0000000001))))
                curr = self.priors[j]*prod
                if curr>m:
                    self.X_prob_[i]=curr
                    m=curr
                    predictions[i]=j
        return predictions

    def score(self, X, y):
        accurate=0
        estimated = self.predict(X)
        for i in range(len(estimated)):
            if estimated[i] == y[i]:
                accurate+=1
                suc_rate = accurate/len(estimated)*100


        return round(suc_rate,4)
```

```python
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()  
    plt.show()

cm1 = confusion_matrix(y_test, gnb.predict(X_test))

plot_confusion_matrix(cm1, unique_labels(y_test),title = "GaussianNB Confusion Matrix")

cm2 = confusion_matrix(y_test, neigh.predict(X_test))

plot_confusion_matrix(cm2, unique_labels(y_test),title = "KNeighbors Confusion Matrix")

cm3 = confusion_matrix(y_test, clfsvc.predict(X_test))

plot_confusion_matrix(cm3, unique_labels(y_test),title = "SVC Confusion Matrix")

```
